```python
import numpy as np
import matplotlib.pyplot as plt

# 1st order eye plant
x = np.zeros(500)  
input = np.zeros(500)
input[100:150] = 1000 #instaintinous input --> pulse input for specific duration
x[0] = 0
dt = 0.001
T1 = 0.15
for t in range(499):
    x[t+1] = x[t] + dt*(-x[t]/T1 +input[t])  # visco-elastic
    

plt.plot(x)
plt.plot(input/100)
plt.show()
```

```python

# 1st order eye plant
x = np.zeros(500)  
input = np.zeros(500)
input[100:150] = 1000 #instaintinous input --> pulse input for specific duration
x[0] = 0
dt = 0.001
T1 = 0.15
ni = np.zeros(500) # we add an integrator and write this and then use this for our model
ni[0] = 0
for t in range(499):
    ni[t+1] = ni[t] + dt*input[t] #simple integrator
    x[t+1] = x[t] + (dt/T1)*(-x[t] +T1*input[t] +ni[t+1])  # visco-elastic
    

plt.plot(x, label = "eye position")
plt.plot(input/100, label = "input")
plt.xlabel("time (ms)")
plt.ylabel("eye position")
plt.legend()
plt.show()
```
